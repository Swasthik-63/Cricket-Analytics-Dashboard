```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
matches = pd.read_csv("../data/matches.csv")
deliveries = pd.read_csv("../data/deliveries.csv")
```

```python
matches["winner"].value_counts()
```

```python
matches["player_of_match"].value_counts()
```

```python
deliveries.groupby("batter")["batsman_runs"].sum()
```

```python
matches["player_of_match"].value_counts().head(10).plot(kind="bar")

plt.title("Top Player of the Match Winners")
plt.xlabel("Player")
plt.ylabel("Awards")
plt.show()
```

```python
matches["venue"].value_counts().head(10)
```

```python
matches["venue"].value_counts().head(10).plot(kind="bar")

plt.title("Top Stadiums by Matches")
plt.xlabel("Venue")
plt.ylabel("Matches")
plt.show()
```

```python
(matches["toss_winner"] == matches["winner"]).value_counts()
```

```python
top_teams = matches["winner"].value_counts().reset_index()
top_teams.columns = ["Team", "Wins"]

top_teams.to_csv("../data/top_teams.csv", index=False)

top_teams.head()
```

```python
top_batsmen = deliveries.groupby("batter")["batsman_runs"].sum().reset_index()

top_batsmen = top_batsmen.sort_values(
    by="batsman_runs",
    ascending=False
)

top_batsmen.to_csv("../data/top_batsmen.csv", index=False)

top_batsmen.head()
```

```python
pom = matches["player_of_match"].value_counts().reset_index()

pom.columns = ["Player", "Awards"]

pom.to_csv("../data/player_of_match.csv", index=False)

pom.head()
```

```python
venues = matches["venue"].value_counts().reset_index()

venues.columns = ["Venue", "Matches"]

venues.to_csv("../data/venues.csv", index=False)

venues.head()
```
